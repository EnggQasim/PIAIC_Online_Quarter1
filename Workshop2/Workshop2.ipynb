# Formating

### .format

```python
name = "Qasim"
fname = "Aslam"
age = 28
gender = "Male"

card = """
PIAIC Admit Card
Name: {}
Father's Name: {}
Age: {} , Gender {}
"""

print(card.format(name, fname, age, gender))
```

```python
name = "Qasim"
fname = "Aslam"
age = 28
gender = "Male"

card = """
PIAIC Admit Card
Name: {1}
Father's Name: {0}
Age: {2} , Gender {3}
"""

print(card.format(name, fname, age, gender))
```

```python
name = "Qasim"
fname = "Aslam"
age = 28
gender = "Male"

card = f"""
PIAIC Admit Card
Name: {name}
Father's Name: {fname}
Age: {age} , Gender {age}
"""

print(card)
```

```python
name = "Qasim"
fname = "Aslam"
age = 28
gender = "Male"

card = f"""
PIAIC Admit Card
Name: %s
Father's Name: %s
Age: %d , Gender %s
"""%(name, fname, age, gender)

print(card)
```

### Slicing

```python
#extract more than one values from any list, tuple or np.array
#list[start,end,step]
#tuple[start,end,step]
#np.array[start:end:step] (":"->seprator)
#start = first index number you have to use to slicing (Include)
#end = first index number you have to use to slicing (exnclude)
#step = sequance

#              0            1           2            3                4
names = ["Inam ul Haq","Dr Noman","Anees Ahmed","Nasir Hussain","Muhamamd Qasim"]
#             -5           -4          -3           -2                -1
#list[start:end:step]
print(names[1:3])# DrNoman --- Nasir Hussain
print(names[1:4])# DrNoman --- Nasir Hussain
print(names[1:4])# DrNoman --- Nasir Hussain
print(names[-4:-1])# DrNoman --- Nasir Hussain
print(names[::2])# DrNoman --- Nasir Hussain
```

```python
numbers = list(range(1,11))
print(numbers)
print(numbers[2:])
print(numbers[2::])
print(numbers[-8:])
```
